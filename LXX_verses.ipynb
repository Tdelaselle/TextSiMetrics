```python
dirpath = "data/LXX-Rahlf-1935/"
```

# Load and verify data

```python
words = pd.read_csv(dirpath+"text_accented.csv",delimiter=',',dtype=str)
refs = pd.read_csv(dirpath+"E-verse.csv",delimiter=',',dtype=str)

# Adding a space for verse reconstruction by joining words
words.word = [str(t)+" " for t in words.word]

display(words.head(5))
display(refs.head(5))
```

```python
LXX = pd.DataFrame(refs.verse)
text = ["".join(list(words.word[(int(refs.ref1[i])-1):(int(refs.ref1[i+1])-1)])) for i in range(refs.shape[0]-1)]
LXX["text"] = text+["".join(list(words.word[(int(refs.ref1[refs.shape[0]-1])-1):]))]
display(LXX)
```

## CSV export

```python
LXX.to_csv('LXX.txt', sep='\t', index=False)
```
