# NT text preprocessing

```python
import pandas as pd
import numpy as np  
# import matplotlib.pyplot as plt 
import openpyxl
import lxml
import nltk

from io import StringIO
from lxml import etree
```

## Load NT file from Proiel 

```python
# Function to clean xml:id attributes
def clean_xml_ids(xml_content):
    # Replace invalid characters in xml:id attributes
    cleaned_content = xml_content.replace('xml:id="RHE_Ps_147_12_c _', 'xml:id="RHE_Ps_147_12_c_')
    return cleaned_content

# Read the XML file content
with open('data/SBLGNT/greek-nt.xml', 'r', encoding='utf-8') as file:
    xml_content = file.read()

# Clean the xml:id attributes
cleaned_xml_content = clean_xml_ids(xml_content)
# display(cleaned_xml_content[0:1000])

# Parse the cleaned XML content
tree = etree.parse('data/SBLGNT/greek-nt.xml')
root = tree.getroot()
```

```python
for elem in root.iter():
    if elem.tag == 'sentence':
        print(elem.tag,elem.attrib)
```

```python
import unicodedata

def remove_greek_accents(text):
    # Normalize the text to decompose characters into base and diacritics
    normalized_text = unicodedata.normalize('NFD', text)
    # Remove diacritical marks (accents)
    stripped_text = ''.join(char for char in normalized_text if not unicodedata.combining(char))
    # Re-normalize to NFC (composed form)
    return unicodedata.normalize('NFC', stripped_text)
```

```python
# Build dataframe listed by sentences, from TEI root

import dis

NT_sentence = pd.DataFrame(columns=['verse_1', 'verse_2', 'token', 'lemma', 'pos'])

token_list = []
lemma_list = []
pos_list = []
v1 = ''
v2 = ''
sentence_id = ''
for elem in root.iter():
    if elem.tag == 'sentence':
        NT_sentence.loc[len(NT_sentence)] = [v1, v2, token_list, lemma_list, pos_list]         
        token_list = []
        lemma_list = []
        pos_list = []
        v1 = ''

    if elem.get('lemma') is not None:
        if v1 == '':
            v1 = elem.get('citation-part')
        lemma_list.append(str(elem.get('lemma')))
        pos_list.append(str(elem.get('part-of-speech')))
        token_list.append(str(elem.get('form')))
    
    if elem.get('citation-part') is not None:
        v2 = elem.get('citation-part')

# Delete first line of NT_sentence
NT_sentence = NT_sentence.drop(NT_sentence.index[0])

# normalize lemmas
NT_sentence['lemma'] = NT_sentence['lemma'].apply(lambda x: [remove_greek_accents(i) for i in x])
NT_sentence['lemma'] = NT_sentence['lemma'].apply(lambda x: [i.lower() for i in x])

display(NT_sentence)
```

```python
# Build dataframe listed by verse, from TEI root

NT = pd.DataFrame(columns=['verse', 'token', 'lemma', 'pos'])

token_list = []
lemma_list = []
pos_list = []
verse = ''
for elem in root.iter():
    if elem.tag == 'token':
        if elem.get('citation-part') != verse:
            NT.loc[len(NT)] = [verse, token_list, lemma_list, pos_list]         
            token_list = []
            lemma_list = []
            pos_list = []
            verse = elem.get('citation-part')
        if elem.get('lemma') is not None:
            lemma_list.append(str(elem.get('lemma')))
            pos_list.append(str(elem.get('part-of-speech')))
            token_list.append(str(elem.get('form')))

# normalize lemmas
NT['lemma'] = NT['lemma'].apply(lambda x: [remove_greek_accents(i) for i in x])
NT['lemma'] = NT['lemma'].apply(lambda x: [i.lower() for i in x])

# Delete first line of NT
NT = NT.drop(NT.index[0])

display(NT)
```

```python
# Function to clean xml:id attributes
def clean_xml_ids(xml_content):
    # Replace invalid characters in xml:id attributes
    cleaned_content = xml_content.replace('xml:id="RHE_Ps_147_12_c _', 'xml:id="RHE_Ps_147_12_c_')
    return cleaned_content

# Read the XML file content
with open('data/UBSGreekNTDic-v1.0-en.xml', 'r', encoding='utf-8') as file:
    xml_content = file.read()

# Clean the xml:id attributes
# cleaned_xml_content = clean_xml_ids(xml_content)
# display(cleaned_xml_content[0:1000])

# Parse the cleaned XML content
tree = etree.parse('data/UBSGreekNTDic-v1.0-en.xml')
root = tree.getroot()
```

```python
import tqdm 

UBS = pd.DataFrame(columns=['lemma', 'domain', 'subdomain','definition'])

lemma_list = []
domain_list = [[]]
subdomain_list = [[]]
definition_list = [[]]

i = -1
j = -1
for elem in tqdm.tqdm(root.iter()):
    if elem.tag == 'Lexicon_Entry':
        lemma_list.append(str(elem.attrib.get('Lemma')))
        domain_list.append([])
        subdomain_list.append([])
        definition_list.append([])
        i +=1
    if elem.tag == 'LEXDomain':
        domain_list[i].append(str(elem.text))
    if elem.tag == 'LEXSubDomain':
        subdomain_list[i].append(str(elem.text))
    if elem.tag == 'DefinitionShort':
        definition_list[i].append(str(elem.text))
        # subdomain_list.append(str(elem.attrib.get('LEXSubDomain')))

# delete last line of domain_list and subdomain_list
domain_list = domain_list[:-1]
subdomain_list = subdomain_list[:-1]
definition_list = definition_list[:-1]

# include lemma_list in UBS
UBS['lemma'] = lemma_list
UBS['domain'] = domain_list
UBS['subdomain'] = subdomain_list
UBS['definition'] = definition_list

# normalize lemmas
UBS['lemma'] = UBS['lemma'].str.lower()
UBS['lemma'] = UBS['lemma'].apply(remove_greek_accents)

display(UBS)
```

```python
def concatenate_all_lists(list_of_lists):
    if len(list_of_lists) < 2:
        if list_of_lists == []:
            return list_of_lists
        else:
            return list_of_lists[0]
    else:
        list_of_lists[0] = list_of_lists[0] + list_of_lists[1]
        del list_of_lists[1]
        return concatenate_all_lists(list_of_lists)    

for i in tqdm.tqdm(range(len(NT['domains']))):
    NT['domains'].values[i] = concatenate_all_lists(NT['domains'].values[i])
    NT['subdomains'].values[i] = concatenate_all_lists(NT['subdomains'].values[i])

    # delete '' from domains and subdomains
    NT['domains'].values[i] = [x for x in NT['domains'].values[i] if x != '']
    NT['subdomains'].values[i] = [x for x in NT['subdomains'].values[i] if x != '']

    NT['domains'].values[i] = list(set(NT['domains'].values[i]))
    NT['subdomains'].values[i] = list(set(NT['subdomains'].values[i]))

for i in tqdm.tqdm(range(len(NT_sentence['domains']))):
    NT_sentence['domains'].values[i] = concatenate_all_lists(NT_sentence['domains'].values[i])
    NT_sentence['subdomains'].values[i] = concatenate_all_lists(NT_sentence['subdomains'].values[i])

    # delete '' from domains and subdomains
    NT_sentence['domains'].values[i] = [x for x in NT_sentence['domains'].values[i] if x != '']
    NT_sentence['subdomains'].values[i] = [x for x in NT_sentence['subdomains'].values[i] if x != '']

    NT_sentence['domains'].values[i] = list(set(NT_sentence['domains'].values[i]))
    NT_sentence['subdomains'].values[i] = list(set(NT_sentence['subdomains'].values[i]))

display(NT)
display(NT_sentence)
```
